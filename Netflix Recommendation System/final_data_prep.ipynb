```python
# Loading the data set
final_data = pd.read_csv("final_netflix_data.csv", index_col=False)
```

```python
# Displaying the data
final_data.head(30)
```

```python
# Referred
# https://stackoverflow.com/questions/37693600/how-to-sort-dataframe-based-on-particular-stringcolumns-using-python-pandas
# Extracting the numbers from the series string
final_data['id_num'] = final_data['show_id'].str.extract('(\d+)', expand=False).astype(int)

# Sorting the dataframe based on the id numbers
final_data = final_data.sort_values('id_num').iloc[:,1:-1]

# Displaying the data
final_data.head(30)
```

```python
# Saving the file as csv
final_data.to_csv("final_sorted_data.csv", index=False)
```
