```python
# Creating a function to select first out of multiple items in a particular column
def select_first(col):
    
    # Creating an empty array to store the first item
    temp = []
    
    # Iterating through the column
    for row in range(0, len(col)):
        
        # Splitting the items in each row
        items = col[row].split(", ")
        
        # Storing the first item in the empty array
        temp.append(items[0])
        
    # Returning the stored items
    return temp
```

```python
# Loading the first data set
show_data = pd.read_csv("show_data.csv", index_col=False)
```

```python
# Dropping unnecessary columns from the data
show_data.drop(labels=["director", "cast"], axis=1, inplace=True)
```

```python
# Displaying the dataframe
show_data.head(30)
```

```python
# Finding the total number of nan in each column
show_data.isna().sum()
```

```python
# Imputing the missing country names with the most occurring country name
show_data.fillna(value=show_data.country.mode()[0], inplace=True)
```

```python
# Finding the total number of nan in each column
show_data.isna().sum()
```

```python
# Retrieving the first country mentioned
country = select_first(show_data.country)

# Adding the country to the dataframe
show_data["country"] = country
```

```python
# Displaying the dataframe
show_data.head(30)
```

```python
# Loading the second data set
show_desc = pd.read_csv("show_description.csv", index_col=False)
```

```python
# Retrieving the genre of the shows
genre = select_first(show_desc.listed_in)

# Adding the genre to the dataframe
show_desc["genre"] = genre
```

```python
# Dropping unnecessary columns from the data
show_desc.drop(labels=["date_added", "listed_in", "description"], axis=1, inplace=True)
```

```python
# Displaying the dataframe
show_desc.head(30)
```

```python
# Finding the total number of nan in each column
show_desc.isna().sum()
```

```python
# Imputing the missing ratings with the most occurring rating
show_desc.fillna(value=show_desc.rating.mode()[0], inplace=True)
```

```python
# Finding the total number of nan in each column
show_desc.isna().sum()
```
